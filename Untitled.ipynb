## Correlation 

```python
# Finding top 10 Correlation for Deafults (Target =1)

corr= Target1.corr()
corrdf= corr.where(np.triu(np.ones(corr.shape),k=1).astype(np.bool))
corrdf= corrdf.unstack().reset_index()
corrdf.columns=['variable1', 'variable2','Correlation']
corrdf.dropna(subset=['Correlation'],inplace= True)
corrdf['Correlation']= round(corrdf['Correlation'],3)
corrdf['Correlation']= abs(corrdf['Correlation'])

corrdf.sort_values(by='Correlation', ascending= False).head(10)
```

```python
# Finding top 10 Correlation for Non-Deafults (Target =0)

corr= Target0.corr()
corrdf= corr.where(np.triu(np.ones(corr.shape),k=1).astype(np.bool))
corrdf= corrdf.unstack().reset_index()
corrdf.columns=['variable1', 'variable2','Correlation']
corrdf.dropna(subset=['Correlation'],inplace= True)
corrdf['Correlation']= round(corrdf['Correlation'],3)
corrdf['Correlation']= abs(corrdf['Correlation'])

corrdf.sort_values(by='Correlation', ascending= False).head(10)
```
